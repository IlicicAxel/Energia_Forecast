```python
import sklearn.metrics
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#Import Pronostico GBA.csv
df = pd.read_csv('Pronostico GBA.csv', index_col=0, dayfirst=True)
df.tail()
```

```python
#Errors 
print("RMSE: %.2f" % sklearn.metrics.mean_squared_error(df['Energia GBA'], df['CAMMESA'], squared = False))
print("MSE: %.2f" % sklearn.metrics.mean_squared_error(df['Energia GBA'], df['CAMMESA']))
print("MAE: %.2f" % sklearn.metrics.mean_absolute_error(df['Energia GBA'], df['CAMMESA']))
```

```python
#Plot time series with FLAML and Energia GBA
plt.figure(figsize=(20,8))
plt.plot(df['Energia GBA'], label='Energia GBA')
plt.plot(df['FLAML'], label='FLAML')
#Date in the x axis format every 24 steps
plt.xticks(np.arange(0, len(df), 24), df.index[np.arange(0, len(df), 24)], rotation=45)
plt.legend()
plt.show()
```

```python
#Plot time series with CAMMESA and Energia GBA
plt.figure(figsize=(20,8))
plt.plot(df['Energia GBA'], label='Energia GBA')
#Change CAMMESA color
plt.plot(df['CAMMESA'], label='CAMMESA', color='red')
#Date in the x axis format every 24 steps
plt.xticks(np.arange(0, len(df), 24), df.index[np.arange(0, len(df), 24)], rotation=45)
plt.legend()
plt.show()
```
